**f)**  
skapar kolumnen Elite, som delar in skolorna i två grupper:  
  
“Yes” om fler än 50 % av studenterna kommer från de 10 % bästa i gymnasiet (Top10perc > 50)  
“No” annars.  
  
och räknar hur många skolor som är “Elite” ("elitstatus") med .value_count().

```python
# Skapa ny variabel 'Elite'
college['Elite'] = pd.cut(
    college['Top10perc'], 
    bins=[0, 50, 100],      # dela in i två grupper: 0-50 och 51-100. 
    include_lowest=True,    # inkludera också den lägsta gränsen (0) 
    labels=['No', 'Yes']
)

# Räkna hur många elituniversitet som finns
print(college['Elite'].value_counts())
```

Går också att göra med np.where(): 

```python
import numpy as np
# Skapa ny variabel 'Elite'

college['Elite'] = np.where(college['Top10perc'] > 50, 'Yes', 'No')
print(college['Elite'].value_counts())  
```

**g)**  
Skapar flera histogram över några kvantitativa variabler (Apps, Accept, Enroll, Outstate).

Undersöker samband mellan olika variabler genom att använda pandas metod .corr.  
Utifrån Pearsons korrelationskoefficient, varje cell visar Pearsons korrelationskoefficient mellan två variabler
https://www.scribbr.com/statistics/pearson-correlation-coefficient/
  
Det vi kan utläsa med hjälp av matrisen är korrelationen mellan två variabler mot varandra.
  
+1 = perfekt positiv linjär korrelation (när en ökar, ökar den andra i samma takt).  
 0 = ingen linjär korrelation.  
-1 = Perfekt negativ linjär korrelation (när en ökar, minskar den andra i samma takt)  
  



```python
college[['Accept', 'Expend', 'Grad.Rate', 'Outstate']].corr()
```

```python
college['AcceptanceRate'] = (college['Accept'] / college['Apps']) * 100
```

Sedan kan vi kolla om det finns samband mellan andel sökande som blivt antagna, kostnad, och andel som tagit examen: 

```python

college[['Grad.Rate', 'Expend', 'AcceptanceRate']].corr()
```

```python
college[['Grad.Rate', 'Outstate', 'AcceptanceRate']].corr()
```

**Grad.Rate vs. Expend** - Medelstarkt samband (0.39). Ju dyrare skolan är, desto högre andel som tar examen.

**AcceptanceRate vs. Expend** - Medelstarkt samband (0.40). Ju dyrare skolan är, desto lägre andel av alla sökande som antas till utbilndningen (därför negativa värden). .
 
Sambandet går att visualisera med hjälp av en scatter plot: 

```python
plt.scatter(college['AcceptanceRate'], college['Expend'])
plt.xlabel("Antagningsfrekvens (%)")
plt.ylabel("Utbildningskostnad (Expend)") 
plt.title("Antagningsfrekvens vs Utbildningskostnad")  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = college['AcceptanceRate']
y = college['Expend']

plt.scatter(x, y, alpha=0.6, label='Skolor')

# Enkel linjär regressionslinje
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', linewidth=2, label='Trendlinje')

plt.xlabel("Antagningsfrekvens (%)")
plt.ylabel("Utbildningskostnad (Expend)")
plt.title("Antagningsfrekvens vs Utbildningskostnad")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
