**f)**  
  
skapar kolumnen Elite, som delar in skolorna i två grupper:  
  
“Yes” om fler än 50 % av studenterna kommer från de 10 % bästa i gymnasiet (Top10perc > 50)  
“No” annars.  
  
och räknar hur många skolor som är “Elite” ("elitstatus") med .value_count().

```python
# Skapa ny variabel 'Elite'
college['Elite'] = pd.cut(
    college['Top10perc'], 
    bins=[0, 50, 100],      # dela in i två grupper: 0-50 och 51-100. 
    include_lowest=True,    # inkludera också den lägsta gränsen (0) 
    labels=['No', 'Yes']
)

# Räkna hur många elituniversitet som finns
print(college['Elite'].value_counts())
```

Går också att göra med np.where(): 

```python
import numpy as np
# Skapa ny variabel 'Elite'
college['Elite'] = np.where(college['Top10perc'] > 50, 'Yes', 'No')
print(college['Elite'].value_counts())  
```

**h)**  
  
Undersöker samband mellan olika variabler genom att bl.a använda pandas metod .corr.  
Utifrån Pearsons korrelationskoefficient, varje cell visar Pearsons korrelationskoefficient(r) mellan två variabler
(https://www.scribbr.com/statistics/pearson-correlation-coefficient/)
  
Det vi kan utläsa med hjälp av matrisen är korrelationen mellan två variabler mot varandra.
  
+1 = perfekt positiv linjär korrelation (när en ökar, ökar den andra i samma takt).  
 0 = ingen linjär korrelation.  
-1 = Perfekt negativ linjär korrelation (när en ökar, minskar den andra i samma takt)  

Generellt kan man tänka att:
- r = 0 = ingen korrelation
- r = 1 = perfekt korrelation
- r = 0-3 = Svag korrelation
- r = 3-5 = Måttligt stark korrelation
- r = >5 = Stark korrelation
(samma gäller även de negativa värdena)

  



```python
college[['Accept', 'Expend', 'Grad.Rate', 'Outstate']].corr()
```

**Grad.Rate och Expend** - Ser en korrelation (r=0.39) mellan antal som tagit examen(Grad.Rate) och kostnad (Expend) för utbildning.  
  
**Grad.Rate och Outstate** - Liknade korrelation (r=0.57) går också att se mellan de som kommer från andra delstater(Outstate) och grad.rate. Där är korrelationen till och med starkare. Kanske för att kostnanderna ofta är högre än Extend?  
Eller för att de flyttat till annan delstat = fler uppoffringar? och därav finns än större motivation till att ta examen?   
  
Korrelationerna tyder på att universitetskostnaden kan påverka studenternas slutbetyg. Om det beror på studenternas motivationen, resurser eller skolans bidragande går dock ej att utläsa. 

Visualiserar sambanden med hjälp av subplots: 

```python
fig, axes = plt.subplots(2, 1, figsize=(12,8), sharex=True)

# 1) Expend - Grad.Rate
axes[0].scatter(college['Expend'], college['Grad.Rate'], alpha=0.6)
axes[0].set_xlabel("Expend ($)")
axes[0].set_ylabel("Graduation Rate (%)")
axes[0].set_title("Correlation between Expend and Grad.Rate")

# 2) Outstate - Grad.Rate
axes[1].scatter(college['Outstate'], college['Grad.Rate'], alpha=0.6)
axes[1].set_xlabel("Out-of-State Tuition ($)")
axes[1].set_ylabel("Graduation Rate (%)")
axes[1].set_title("Correlation btween Outstate and Grad.Rate")

plt.show()
```

 Jag vill kunna se om det finns ett samband i hur eftertraktad skolan är och hur stor andel som tar examen. Då alla skolor är av olika storlek kan vi inte mäta det genom att jämföra Grad.Rate med App eller Accept bara. Men vi kan ta ut hur stor andel av alla som sökt till varje skola som blivit antagna och på så sätt få en bild av hur attraktivt varje universitet är.  
  
För att få ut antal antagna av alla sökande skapar jag en ny variablel/kategori; "AcceptanceRate", som visar hur hög antagningsgraden är i procent och ser hur fördelningen ser ut med hjälp av decribe () och ett Histogram. Vi börjar med den nya variabeln genom att dela "Accept" - De som fått erbjudande om plats, med "App" - Alla som sökt och multiplicerar med 100:

```python
college['AcceptanceRate'] = (college['Accept'] / college['Apps']) * 100
college['AcceptanceRate'].describe()
```

```python
plt.hist(college['AcceptanceRate'], bins=50)
plt.title("Histogram of Acceptance Rate (%)")
plt.xlabel("Acceptance Rate (%)")
plt.ylabel("Number of Colleges")
plt.show()
```

Sedan kan vi kolla om det finns samband mellan AcceptanceRate och andra variabler, exempelvis kostnad och andel som tagit examen: 

```python

college[['Grad.Rate', 'Expend', 'AcceptanceRate']].corr()
```

**AcceptanceRate och Grad.Rate** - Samband finns, men det är relativt svagt(r=0.29)

**AcceptanceRate vs. Expend** - Måttligt starkt samband (r=0.40). Ju dyrare skolan är, desto lägre andel av alla sökande antas till utbilndningen (därför negativa värden). .
Sambandet går att visualisera med hjälp av en scatter plot med en enkel linjär regressionslinje: 

```python
x = college['AcceptanceRate']
y = college['Expend']

plt.scatter(x, y, alpha=0.6, label='Skolor')

m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color='red', label='Trendlinje')

plt.xlabel("AcceptanceRate (%)")
plt.ylabel("Expend ($)") 
plt.title("Correlation between Acceptance Rate and Expend") 
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

Kan också kolla om det finns någon korrelation mellan Acceptance Rate och Elite:

```python
college.boxplot(column="AcceptanceRate", by="Elite", figsize=(6,4))
plt.title(" Acceptance rate in relation to elitstatus")
plt.suptitle("")
plt.xlabel("Elite")
plt.ylabel("Acceptance Rate (%)")
plt.show()
```

Resultatet av ovan jämförelse visar att det finns ett samband mellan andel antagna och elitstatus. 
  
Skolor som har elitstatus tar in färre andel av alla som ansöker till skolan, än skolor som inte har elitstatus. Kanske då de anses mer eftertraktade (ökat tryck i ansökningar) på grund av sitt elitstatus.
